# <font size="6" face="verdana" color="blue">What Is Regression?</font>

Regression searches for **relationships** among variables.


For example, you can observe several employees of some company and try to understand how their salaries depend on the features, such as experience, level of education, role, the city they work in, and so on.

This is a regression problem where data related to each employee represent one **observation**. The presumption is that the experience, education, role, and city are the **independent features**, while the salary depends on them.

Generally, in regression analysis, you usually consider some phenomenon of interest and have a number of observations. Each observation has two or more features. Following the assumption that (at least) one of the features depends on the others, you try to establish a relation among them.

In other words, **you need to find a function that maps some features or variables to others sufficiently well.**

The dependent features are called the **dependent variables, outputs, or responses**.

The independent features are called the **independent variables, inputs, or predictors**.

Regression problems usually have one continuous and unbounded dependent variable. The inputs, however, can be continuous, discrete, or even categorical data such as gender, nationality, brand, and so on.

It is a common practice to denote the outputs with $y$ and inputs with $x$. If there are two or more independent variables, they can be represented as the vector $x = (x_1,...,x_n)$, where $n$ is the number of inputs.

# <font size="6" face="verdana" color="blue">When Do You Need Regression?</font>

* Typically, you need regression to answer **whether and how some phenomenon influences the other or how several variables are related**. For example, you can use it to determine if and to what extent the experience or gender impact salaries.


* Regression is also useful when you want to **forecast a response using a new set of predictors**. For example, you could try to predict electricity consumption of a household for the next hour given the outdoor temperature, time of day, and the number of residents in that household.

# <font size="6" face="verdana" color="blue">Linear Regression</font>
    
Linear regression is a technique used to describe the relationship between the dependent variable and one or more independent variables.

When implementing linear regression of some dependent variable ùë¶ on the set of independent variables $x = (x_1,..., x_n)$, where $n$ is the number of predictors, you assume a linear relationship between $y$ and $x$: 

$$ y = \beta_0 + \beta_1x_1 + ... + \beta_nx_n + \epsilon$$

<div class="alert alert-block alert-warning">  
$$ y = \beta_0 + \beta_1x_1 + ... + \beta_nx_n + \epsilon$$ 
</div>

This equation should capture the dependencies between the inputs and output sufficiently well.

This equation is the regression equation. $\beta_0, \beta_1, ... ,\beta_n $ are the regression coefficients, and $\epsilon$ is the random error.

Linear regression calculates the estimators of the regression coefficients or simply the predicted weights.

<img src='images/fig-lin-reg.jpg' />

**Intercept** - The value of $\beta_0 $ ¬†also called the intercept, shows the point where the estimated regression line crosses the $ùë¶$ axis. It is the value of the estimated response of $y$ for $x=0$. 

**Slope** - The value of $\beta_1$ determines the slope of the estimated regression line.

**Residual** - $\epsilon$ represents the residual value, i.e. the difference between actual and predicted values. They are the distances between the blue circles and red squares. When we implement linear regression, we are actually trying to minimize these distances and make the red squares as close to the predefined blue circles as possible.



#### So, Linear Regression calculates an equation that minimizes the distance between the fitted line and all of the data points.


The estimated or predicted response, $ \hat{y}_i$, for each observation $i = 1,..., n$, should be as close as possible to the corresponding actual response $y_i$. The differences ( $ \hat{y}_i - y_i $ ) for all observations $i = 1,..., n$, are called the **residuals**. Regression is about determining the best predicted weights, that is the weights corresponding to the smallest residuals.

# <font size="6" face="verdana" color="blue">The 3 Sums of Squares</font>

There are three terms we must define. The **sum of squares total (SST)**, the **sum of squares regression (SSR)**, and the **sum of squares error (SSE)**.

* ## Sum of Squares Total (SST) / (TSS) : Difference between actual value and mean
The sum of squares total, denoted SST, is the squared differences between the observed dependent variable and its mean. You can think of this as the dispersion of the observed variables around the mean ‚Äì much like the variance in descriptive statistics.

$$ SST = \sum\limits_{i=1}^{n}(y_i - \bar{y}_i)^2 ¬†$$

<div class="alert alert-block alert-warning">  
$$ SST = \sum\limits_{i=1}^{n}(y_i - \bar{y}_i)^2 ¬†$$
</div>

¬† ¬† It measures the total variability of the dataset.

* ## Sum of Squares Regression (SSR) / (ESS) : Difference between predicted value and mean
It is the sum of the differences between the predicted value and the mean of the dependent variable. Think of it as a measure that describes how well our line fits the data.

$$ SSR = \sum\limits_{i=1}^{n}(\hat{y}_i - \bar{y}_i)^2 ¬†$$

¬† ¬† It is a measure that describes how well your line fits the data.

* ## Sum of Squares Error (SSR) / (RSS) : Difference between predicted value and actual value
The error is the difference between the observed value and the predicted value.

$$ SSE = \sum\limits_{i=1}^{n}(y_i - \hat{y}_i)^2 ¬†$$

<div class="alert alert-block alert-warning">  
<h3> $$ SSE = \sum\limits_{i=1}^{n}(y_i - \hat{y}_i)^2 ¬†$$ </h3>
</div>

¬† ¬† It measures the unexplained variability by the regression. The smaller the error the better the estimation power of the regression.

Mathematically,
## $$ SST = SSR + SSE $$

<div class="alert alert-block alert-warning">  
<h2> $$ SST = SSR + SSE $$ </h2>
</div>


## Interpretation:
* The total variability of the data set is equal to the variability explained by the regression line plus the unexplained variability, known as error.
* If the value of SSR is equal to the SST, it means our regression model captures all the observed variability and is perfect.
* A lower error will cause a better regression.
* Conversely a higher error will cause a less powerful regression.
* SSE and SSR are inversely proportional.
* Our goal is to calculate the optimal values of the predicted weights $\beta_0, \beta_1, ... ,\beta_n $ ¬†that minimize SSR and determine the estimated regression function.¬†

# <font size="6" face="verdana" color="blue">Underfitting and Overfitting</font>

The cause of poor performance in machine learning is either overfitting or underfitting the data.


**Overfitting** - When predicted model gives very low error on the training dataset, however, the same model gives a high error for the test dataset.

Overfitting is the case where the overall cost is really small, but the generalization of the model is unreliable. This is due to the model learning ‚Äútoo much‚Äù from the training data set.


**underfitting** - Underfitting means the model gives hight error for both train and test dataset. ¬†

Underfitting is the case where the model has ‚Äú not learned enough‚Äù from the training data, resulting in low generalization and unreliable predictions.

A model is said to be a good machine learning model if it generalizes any new input data from the problem domain in a proper way. This helps us to make predictions in the future data, that data model has never seen.

Now, suppose we want to check how well our machine learning model learns and generalizes to the new data. For that, we have overfitting and underfitting, which are majorly responsible for the poor performances of the machine learning algorithms.

## Overfitting:
A statistical model is said to be overfitted when we train it with a lot of data. When a model gets trained with so much data, it starts learning from the noise and inaccurate data entries in data set, then the model gives predicts incorrectly on unseen/test dataset. ¬†When applied to known data, such models usually yield high $R^2$. However, they often have significantly lower $R^2$ when used with new data.

**It usually happens:** 
* When the model start learning from the noise and inaccurate data entries in the data set.


**How to avoid Overfitting:**
The commonly used methodologies are:

* **Cross-Validation:** A standard way to find out-of-sample prediction error is to use 5-fold cross-validation.
* **Early Stopping:** Its rules provide us with the guidance as to how many iterations can be run before the learner begins to over-fit.
* **Pruning:** Pruning is extensively used while building-related models. It simply removes the nodes which add little predictive power for the problem in hand.
* **Regularization:** It introduces a cost term for bringing in more features with the objective function. Hence it tries to push the coefficients for many variables to zero and hence reduce cost term.
¬† ¬† ¬† ¬† They are three types of regularization technique to overcome overfitting.
¬† ¬† ¬† ¬† ¬† ¬† ¬†a) L1 regularization (also called Lasso regularization / panelization.)
¬† ¬† ¬† ¬† ¬† ¬† ¬†b) L2 regularization (also called Ridege regularization/ penalization.)
¬† ¬† ¬† ¬† ¬† ¬† ¬†c) Elastic net


## Underfitting:
A statistical model or a machine learning algorithm is said to have underfitting when it cannot capture the underlying trend of the data. Underfitting destroys the accuracy of our machine learning model. Its occurrence simply means that our model or the algorithm does not fit the data well enough. ¬†It often yields a low $R^2$ with known data and bad generalization capabilities when applied with new data.

**It usually happens:** 
* when we have fewer data to build an accurate model and 
* also when we try to build a linear model with a non-linear data. 

In such cases, the rules of the machine learning model are too easy and flexible to be applied on such minimal data and therefore the model will probably make a lot of wrong predictions. Underfitting can be avoided by using more data and also reducing the features by feature selection.


### Good Fit in a Statistical Model:

Ideally, the case when the model makes the predictions with 0 error, is said to have a good fit on the data. This situation is achievable at a spot between overfitting and underfitting. In order to understand it, we will have to look at the performance of our model with the passage of time, while it is learning from the training dataset.

With the passage of time, our model will keep on learning and thus the error for the model on the training and testing data will keep on decreasing. If it will learn for too long, the model will become more prone to overfitting due to the presence of noise and less useful details. Hence the performance of our model will decrease. In order to get a good fit, we will stop at a point just before where the error starts increasing. At this point, the model is said to have good skills on training dataset as well as our unseen testing dataset.



# <font size="6" face="verdana" color="blue">Coefficient of determination or R-Squared ($R^2$)</font>

R-squared is a statistical measure of how close the data are to the fitted regression line. R-squared (R2) is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model.

The most general definition of the coefficient of determination is


$$ R^2 =  {SSR \over SST} $$

$$ R^2 =  {SSR \over SSR+SSE} $$

$$ R^2 = 1- {SSE \over SST} $$

$$ R^2 = {VariabilityExplainedByRegression \over TotalVariability} $$

$$ 0 \leq R^2 \leq 1 $$

* if there is no error in model (SSE = 0), then $R^2$ becomes 1.

### R-squared is always between 0 and 100%:

* 0% indicates that the model explains none of the variability of the response data around its mean.
* 100% indicates that the model explains all the variability of the response data around its mean.
* In general, the higher the R-squared, the better the model fits your data.
* The more factors you include in your regression model, the higher the $R^2$.
* $R^2$ explains the degree to which your input variables explains the variation of your predicted/output variable.
* So if $R^2$ is 0.8, it means 80% of the variance in the output variable is explained by the input variables.


In simple terms, higher the $R^2$, the more variation is explained by your input variables and hence better is your model.

However, the problem with $R^2$ is that it will either stay the same or increase with addition of more variables, even if they do not have any relationship with the output variable.This is where **Adjusted $R^2$** comes to help. $R^{-2}$ penalizes you for adding variable which do not improve your existing model.

In case, you only have one input variable, $R^{2}$ and $R^{-2}$ whould be exactly same.

Typically, the more non-significant variables you add into the model, the gap is $R^{2}$ and $R^{-2}$ increases.

# <font size="6" face="verdana" color="blue">Adjusted R-Squared ($R^{-2}$)</font>

R-Squared only works as intended in a simple linear regression model with one explanatory variable. With a multiple regression made up of several independent variables, the R-Squared must be adjusted. The adjusted R-squared compares the descriptive power of regression models that include diverse numbers of predictors. Every predictor added to a model increases R-squared and never decreases it. Thus, a model with more terms may seem to have a better fit just for the fact that it has more terms, while the adjusted R-squared compensates for the addition of variables and only increases if the new term enhances the model above what would be obtained by probability and decreases when a predictor enhances the model less than what is predicted by chance. In an overfitting condition, an incorrectly high value of R-squared is obtained, even when the model actually has a decreased ability to predict. This is not the case with the adjusted R-squared.

$$ R^{-2} = 1-{(1-R^{2})(n-1) \over {(n-(1+k))}} $$ 

* The $R^{2}$ measures how much of the total variability is explained by our model.
* Multiple regression is always better than simple ones, as, with each additional variabile you add, the explanatory power may only increase or stay the same.

$$ R^{-2} < R^{2}$$

The adjusted R-Squared is always less than R-Squared as adjusted R squared penalizes excessive use of variables.

### Some problems with R-Squared
* Problem 1 -Every time you add a predictor to a model, the $R^{2}$ increases, even if due to chance alone, it never decreases. Consequently, a model with more terms may appear to have a better fit simply because it has more terms.

* Problem 2 - If a model has too many predictors and higher-order polynomials, it begins to model the random noise in the data. It produces misleadingly high $R^{2}$ values and lessened ability to make predictions.


$R^{2}$ does not indicate whether a regression model is adequate. You can have a low $R^{2}$ value for a good model or a high $R^{2}$ value for a model that does not fit the data.

### What is Adjusted R-Squared?
The adjusted R-Squared compares the explanatory power of the regression model that contain a different number of predictors.

Suppose you compare a five predictor model with a higher $R^{2}$ to a one predictor model. Does the five predictor model have a higher $R^{2}$ because it's better? Or is the $R^{2}$ higher because it has more predictors? Simply compare the $R^{-2}$ values to find out.

**Important:-** $R^{-2}$ is a modified version of $R^{2}$ that has been adjusted for the number of predictors in the model. The $R^{-2}$ increases only if the new term improves the model more than would be expected by chance. The $R^{-2}$ can be negative, but it is usually not. It is always lower than $R^{2}$.

# <font size="6" face="verdana" color="blue">Ordinary least squares (OLS)</font>

ordinary least squares (OLS) is a type of linear least-squares method for estimating the unknown parameters in a linear regression model. OLS chooses the parameters of a linear function of a set of explanatory variables by the principle of least squares: minimizing the sum of the squares of the differences between the observed dependent variable in the given dataset and those predicted by the linear function (SSE).

Least Squares stands for the minimum sum of square errors or $minSSE$. The OLS method corresponds to minimizing the sum of square differences between the observed and predicted values. 

Ordinary least squares is the most common method to estimate the linear regression equation. 
¬†
So this method aims to find the line which minimizes the sum of the squared errors. 
¬†

# <font size="6" face="verdana" color="blue">Assumptions for Linear Regression Analysis</font>


* **Linearity**: The relationship between X and the mean of Y is linear.


* **Homoscedasticity**: Errors should have constant variance.


* **Independence**: Observations are independent of each other.


* **Normality**: Dependent and independed variables should be normally distributed. Also Errors should be normally distributed.


* **No multicollinearity**: No correlation between between independent variables.


* **No auto-correlation**: No correlation between error terms.


* **No Endogeneity**: No correlation between error and independent variable.

# <font size="6" face="verdana" color="blue">Correlation Matrix</font>

The correlation matrix is a square matrix that contains the Pearson product-moment
correlation coefficients (often abbreviated as Pearson's r), which measure the linear
dependence between pairs of features. The correlation coefficients are bounded
to the range -1 and 1. Two features have a perfect positive correlation if $r=1$, no
correlation if $r=0$ and a perfect negative correlation if $r=-1$ respectively. As
mentioned previously, Pearson's correlation coefficient can simply be calculated as
the covariance between two features $x$ and $y$ (numerator) divided by the product
of their standard deviations (denominator):  

\begin{equation}
r=\frac{\sum_{i=1}^n[(X^{(i)}-\mu_{x}) (y^{(i)}-\mu_{y})]}{\sqrt{\sum_{i=1}^n(X^{(i)}-\mu_{x})^2}\sqrt{\sum_{i=1}^n(y^{(i)}-\mu_{y})^2}}  = \frac{\sigma_{xy}}{\sigma_{x}\sigma{y}}
\end{equation}

* This value can be calculated only between two numeric values
* Correlation between [-1,0) means inversely proportional, the scatter plot will show a downward trend
* Correlation between (0,1] means directly proportional, the scatter plot will show a upward trend
* Correlation near {0} means No relationship, the scatter plot will show no clear trend.
* If Correlation value between two variables is > 0.5 in magnitude, it indicates good relationship the sign does not matter
* We observe the correlations between Target variable and all other predictor variables(s) to check which columns/features/predictors are actually related to the target variable in question

## $R^{-2}$  Adjusted R Squared

$$ R^{-2} = 1-{(1-R^{2})(n-1) \over {(n-(1+k))}} $$ 

adj_rsquared = 1 - (1 - R2) * ((n - 1)/(n-p-1)) 


```python
adj_rsquared = 1 - (1 - r_sq) * ((500 - 1)/(500-8-1))
print('Adjusted R Squared : ',adj_rsquared)
```
